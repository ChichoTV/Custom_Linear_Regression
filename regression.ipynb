```python
#regression function, returning a TUPLE containing:
#First, a linear function using the method of least squares to best fit data
#Second, a list containing the coefficients of the polynomial (the slope and y-intercept respectively in the list)
#yes, there are way better functions out there, this is just for my own practice but feel free to toy with it.
#the function accepts lists as parameters
#when setting a variable equal to the function you must define it as a tuple as seen below. The first element will hold the 
#equation of the line itself while the second will hold the list of coefficeints.
#numpy and scipy.linalg (or ) are necessary libraries
```

```python
%matplotlib notebook
import numpy as np
import numpy.linalg as la
import matplotlib.pyplot as plt
```

```python
def linear_reg(x,y):
    xarr=np.array(x)
    yarr=np.array(y)
    xarr_squared=np.square(xarr)
    xone_one=xarr_squared.sum()
    xone_two=xarr.sum()
    xtwo_one=xarr.sum()
    xtwo_two=np.ones(len(xarr)).sum()
    X=np.array([[xone_one,xone_two],[xtwo_one,xtwo_two]])
    temp=np.zeros(len(xarr))
    for i in range(len(xarr)):
        temp[i]=xarr[i]*yarr[i]
    yone=temp.sum()
    ytwo=yarr.sum()
    Y=np.array([yone,ytwo])
    a=la.solve(X,Y)
    f=lambda z:z*a[0]+a[1]
    return (f,a)
```

```python
temp=[14.2, 16.4, 11.9, 15.2, 18.5, 22.1, 19.4, 25.1, 23.4, 18.1, 22.6, 17.2]
sales = [215, 325, 185, 332, 406, 522, 412, 614, 544, 421, 445, 408]
temp_array=np.array(temp)
```

```python
func,coeff=linear_reg(temp,sales)
```

```python
plt.plot(temp,func(temp_array),color="red",label="Linear Regression")
plt.scatter(temp,sales,label="Data")
plt.legend(loc="lower right")
plt.title("Ice Cream Sales VS Temperature")
plt.xlabel("Temperature in Celsius")
plt.ylabel("Sales of Ice Cream")
```
